```python
# Count the number of movies and TV shows
movie_count = df[df['type'] == 'Movie'].shape[0]
tv_show_count = df[df['type'] == 'TV Show'].shape[0]

# Print the results
print(f"Number of Movies: {movie_count}")
print(f"Number of TV Shows: {tv_show_count}")

```

```python
data = [movie_count, tv_show_count]
labels = ['Movie', 'TV Show']

plt.figure(figsize=(8, 6))
plt.pie(data, labels=labels, autopct="%1.1f%%", startangle=140, colors=['b', 'm'])  # Adjust options as desired
plt.title('Comparison of Movies and TV Shows (Pie Chart)')

plt.tight_layout()
plt.show()
```
